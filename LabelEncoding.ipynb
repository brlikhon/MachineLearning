```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('/content/weather.numeric.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l = LabelEncoder()
df['New_Play'] = l.fit_transform(df.play)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh = OneHotEncoder(sparse=False,handle_unknown='error',drop='first')
```

```python
n = oh.fit_transform(df[['outlook']])
n
```

```python
dfn = pd.DataFrame(n)
dfn
```

```python
df = df.join(dfn)
df
```
